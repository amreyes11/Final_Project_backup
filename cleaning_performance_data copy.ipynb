```python
#import dependencies
import pandas as pd
```

```python
#Read in povery data from CSV
performance_df = pd.read_csv("data/california_hospitals_performance_ratings.csv", header=0, encoding='cp1252')
performance_df.head()
```

```python
#Remove unnecessary columns
clean_df=performance_df.drop(columns=["OSHPDID","system", "Type of Report"])
clean_df.head()
```

```python
#Count performance measures by year, county, and hospital ratings
counties_df=clean_df.groupby(["Year","County","Hospital Ratings"]).count()["Performance Measure"]
counties_df.head(10)
```

```python
#hospitals_df=clean_df.groupby(["Year","County","Hospital"]).count()["OSHPDID"]
#hospitals_df.head(30)
#hospitals_df=clean_df.groupby(["Year", "County"])['Hospital'].value_counts()
#hospitals_df.head(30)
```

```python
#Count adverse events by year, county, and hospital ratings
adverse_df=clean_df.groupby(["Year","County","Hospital Ratings"]).sum()["# of Adverse Events"]
adverse_df.head(10)
```

```python
#Count number of cases by year, county, and hospital ratings
cases_df=clean_df.groupby(["Year","County","Hospital Ratings"]).sum()["# of Cases"]
cases_df.head(10)
```

```python
#Create new df
performance_summary_df = pd.DataFrame()

performance_summary_df["Hospital Ratings Total"] = counties_df
performance_summary_df["Total Number of Adverse Events"]= adverse_df
performance_summary_df["Total Number of Cases"]= cases_df

performance_summary_df
```

```python
performance_summary_df.to_csv('performance_summary.csv')
```
