```python
#import dependencies
import pandas as pd
```

```python
#Read in performance data from CSV
performance_df = pd.read_csv("data/small_area_income_and_poverty_estimates.csv", header=0)
performance_df.head()
```

```python
#Remove unnecessary columns
clean_df=performance_df.drop(columns=["ID","Poverty Universe", "90% Confidence Interval","90% Confidence Interval.1"])
clean_df.head()
```

```python
#Remove unnecessary rows
clean2_df=clean_df[clean_df['Name'].str.contains('United States')==False]
clean2_df=clean2_df[clean2_df['Name'].str.contains('California')==False]
clean2_df.head()
```

```python
clean_name= [i.replace(' County','') for i in clean2_df["Name"].to_list()]
clean2_df['clean_name']=clean_name
clean2_df.head()
```

```python
#Remove unnecessary columns
clean2_df=clean2_df.drop(columns=["Name"])
clean2_df.head()
```

```python
#Change "name" column to "county"
clean2_df.rename(columns={'clean_name': 'County'}, inplace=True)
clean2_df.head()
```

```python
clean2_df=clean2_df.loc[:,['Year','County','Number in Poverty','Percent in Poverty']]
```

```python
clean2_df.head()
```

```python
clean2_df.to_csv('poverty_summary.csv')
```
